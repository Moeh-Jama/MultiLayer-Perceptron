```python
from multi_layer_perceptron import MultiLayerPerceptron
from main import cross_entropy,squared_error,Example
from main import readDigitRecognition
import random
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import normalize
import connectionistC_q3 as q3
import time
```

```python
def display_error_rates(error_rates, learning_rate=0.1):
    plt.plot(error_rates)
    plt.ylabel('error rates')
    plt.xlabel("epoch")
    plt.suptitle("Learning Rate:={}".format(learning_rate), fontsize=20)
    plt.show()
```

```python
def learn_generally(nn, training_examples, learning_rate=0.1, max_epochs=100):
    
    inputs = []
    outputs = []
    
    for example in training_examples:
        inputs.append(example.input)
        outputs.append(example.output)
    inputs = np.array((inputs), dtype=float)
    outputs = np.array((outputs), dtype=float)
    
    error_rates = nn.train(inputs, outputs, max_epochs, learning_rate=learning_rate)
    display_error_rates(error_rates,learning_rate)
    
    return nn

def validate_learning_CLASSIFICATION(nn, test_examples):
    
    correct_count = 0
    for example in test_examples:
        X = example.input
        target = example.output
        
        predicted = list(nn.forward(X))
        predicted = predicted.index(max(predicted))
        target = target.index(max(target))
        if target==predicted:
            correct_count +=1
    acc = correct_count / len(test_examples)
    print('accuracy is: {}%'.format( round(acc*100,2) ))
        
        
def validate_learning_REGRESSION(nn, test_examples):
    error = 0
    
    for example in test_examples:
        
        X = example.input
        target = example.output
        
        predicted = nn.forward(X)[0]
        error = error + squared_error(target, predicted)

    print('mean squared error {}'.format((error/len(test_examples))))
    
```

```python
def displayWeightMatrix(Msize, matrix, msg):
    # This visualisation can be found at
    # https://stackoverflow.com/questions/33828780/matplotlib-display-array-values-with-imshow
    # it was slightly modified with 
    # https://stackoverflow.com/questions/33282368/plotting-a-2d-heatmap-with-matplotlib
    
    
    # matrix = normalize(matrix, axis=1, norm='l1')
    size =Msize
    # data = np.arange(size * size).reshape((size, size))
    x_start = 3.0
    x_end = 9.0
    y_start = 6.0
    y_end = 12.0
    extent = [x_start, x_end, y_start, y_end]
    fig = plt.figure(figsize=(16, 12))
    ax = fig.add_subplot(111)

    im = ax.imshow(matrix,extent=extent, cmap='viridis', origin='lower',interpolation='nearest' )


    jump_x = (x_end - x_start) / (2.0 * size)
    jump_y = (y_end - y_start) / (2.0 * size)
    x_positions = np.linspace(start=x_start, stop=x_end, num=size, endpoint=False)
    y_positions = np.linspace(start=y_start, stop=y_end, num=size, endpoint=False)


    for y_index, y in enumerate(y_positions):
        for x_index, x in enumerate(x_positions):
            # label = values[y_index, x_index]
#             label = "({},{})\n{}".format(y_index,x_index, round(matrix[y_index][x_index],4))
            # label = "({},{}).format(y_index,x_index)
            label = ""
            # print('label:',label)
            text_x = x + jump_x
            text_y = y + jump_y
            ax.text(text_x, text_y, label, color='black', ha='center', va='center')

    fig.colorbar(im)
    plt.suptitle(msg, fontsize=20)
    plt.show()
```

```python
def get_SIN():

    nn = MultiLayerPerceptron(4,5,1,squared_error)
    return nn

def get_XOR():
    nn = MultiLayerPerceptron(2,3,1,squared_error)
    return nn

def get_LEARN_LETTERS():
    nn = MultiLayerPerceptron(16,15,26,cross_entropy, batch=50)
    return nn
```

```python

def get_train_test_sets(example_list, lim):
    train = []
    test =[]
    lim = int(len(example_list) * (2/3))
    for example in example_list[:lim]:
        train.append(example)
    
    for example in example_list[lim:]:
        test.append(example)
    
    return train, test


def get_letter_dataset():
    vectors, outputs, label_mapping, map_label = readDigitRecognition()
    example_list = []
    train = []
    test = []
    for i in range(len(vectors)):
        example = Example(vectors[i],outputs[i])
        example_list.append(example)
    
    # The dataset is already randomly sorted, with unlikely biases.
    lim = int(len(example_list) * (2/3))
    train, test = get_train_test_sets(example_list, lim)
    
    
    return train, test, label_mapping, map_label
    
def get_XOR_dataset():
    vectors = [[0,0],[0,1],[1,0],[1,1]]
    outputs = [0,1,1,0]
    example_list = []
    for i in range(len(vectors)):
        example = Example(vectors[i],outputs[i])
        example_list.append(example)
    
    return example_list ,example_list

def get_SIN_dataset():
    vectors,outputs = q3.getVector_List()
    example_list = []
    for i in range(len(vectors)):
        example = Example(vectors[i],outputs[i])
        example_list.append(example)
    
    random.shuffle(example_list)
    lim = int(len(example_list) * (2/3))
    train, test = get_train_test_sets(example_list, lim)
    
    return train, test
    
    
    
    
```

```python
def experiement_XOR():

    nn = get_XOR()
    
    training_examples, test_examples = get_XOR_dataset()


    nn.display=True
    nn.output_steps=1000
    nn = learn_generally(nn, training_examples, max_epochs=1000, learning_rate=0.25)
    mse = validate_learning_REGRESSION(nn, test_examples)

    return nn
```

```python
def experiement_SIN():    
    training_examples, test_examples = get_SIN_dataset()
    
    nn = get_SIN()
    nn.display=False
    nn = learn_generally(nn, training_examples, max_epochs=1000, learning_rate=0.125)
    mse = validate_learning_REGRESSION(nn, test_examples)
    return nn
```

```python
def experiment_letter_rec(learning_rate=0.1, max_epochs=100):
    nn = get_LEARN_LETTERS()
    training_examples, test_examples, lm, ml = get_letter_dataset()

    nn.display=True
    nn.output_steps=5
    nn = learn_generally(nn, training_examples, max_epochs=max_epochs, learning_rate=learning_rate)
    validate_learning_CLASSIFICATION(nn, test_examples)

    return nn
```
